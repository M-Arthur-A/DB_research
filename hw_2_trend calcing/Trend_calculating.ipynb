```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.float_format', '{:.2f}'.format)
np.set_printoptions(formatter={'float_kind': "{:.2f}".format})
%config InlineBackend.figure_format = 'svg'
```

```python
path = '/home/arthur/Project/GeekBrains/DB_research/!ADDS/hw_1/orders_20190822.csv'
df = pd.read_csv(path, sep=';')
df
```

```python
# задание соответствующих форматов столбцам
df['o_date'] = pd.to_datetime(df['o_date'], format='%d.%m.%Y')
df['year'] = df['o_date'].dt.year
df['price'] = df['price'].str.replace(',', '.').astype('float')
df['month'] = df['o_date'].dt.month
```

```python
season_pivot = df.pivot_table(index=['year', 'month'], values='price', aggfunc=np.sum)
season_pivot = season_pivot.reset_index()
df = season_pivot.copy()
df
```

## (1) Расчет значение тренда

```python
# использование аналога excel-функции ЛИНЕЙН (LINEST) в numpy
Y = np.arange(1, df.shape[0]+1)
X = df['price'].to_numpy()
linest = np.polyfit(Y, X, 1)
b = linest[0]
a = linest[1]
b, a
```

```python
# рассчитываем значения тренда
df['id'] = Y
df['trend'] = df['id'] * b + a
df
```

## (2) Определение коэффициентов сезонности

```python
# рассчитываем отклонение фактических значений от значений тренда
df['trend_delta'] = df['price'] / df['trend']
df
```

```python
# Среднее отклонение для каждого месяца
avg_month_trend_delta = df.pivot_table(index='month', values='price',aggfunc=np.mean)
avg_month_trend_delta
```

```python
common_season_index = avg_month_trend_delta.mean().squeeze()
common_season_index
```

```python
avg_month_trend_delta['season_index'] = avg_month_trend_delta['price'] / common_season_index
avg_month_trend_delta
```

## (3) Прогнозирование продаж

```python
# задаем период, на который мы хотим рассчитать прогноз.
future = np.arange(df.shape[0]+1, df.shape[0]+1+12)
df_future = pd.DataFrame(index=future)

# рассчитываем значения трена для будущих периодов
df_future['f_trend'] = df_future.index * b + a
df_future['month'] = np.arange(1,13)

# подтягиваем коэффициены сезонности
df_future = df_future.merge(avg_month_trend_delta, left_on='month', right_index=True)
df_future
```

```python
# Рассчитываем прогноз
df_future['f_price'] = df_future['f_trend'] * df_future['season_index']
df_future
```
