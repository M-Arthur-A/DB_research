# (3) По годам посчитать средний чек, среднее кол-во заказов на пользователя, сделать вывод , как изменялись это показатели Год от года.

```python
# задание соответствующих форматов столбцам
df['year'] = df['o_date'].dt.year
df['price'] = df['price'].str.replace(',', '').astype('int64')
```

```python
# средний чек по годам
# считаем медиану, чтобы исключить выбросы (а они тут есть, в т.ч. отрицательные цены)
df.pivot_table(index=['year'], values=['price'], aggfunc=np.median)
```

```python
# среднее число заказов на пользователя
df.pivot_table(index=['user_id'], values=['id_o'], aggfunc='count').mean().squeeze()
```

```python
# среднее число заказов на пользователя по годам
user_mean_o = df.pivot_table(index=['year', 'user_id'], values=['id_o'], aggfunc='count')
user_mean_o = user_mean_o.reset_index()
user_mean_o.pivot_table(index=['year'], values=['id_o'], aggfunc=np.mean)
```

**Вывод**  
При падении среднего числа заказов на одного покупателя, средний чек увеличивался.  
Возможно падение спроса на продукцию на фоне повышения цен (возможно инфляция, либо изменился асортимент на более дорогой, либо просто цены подняли, либо изменилась структура спроса в части предпочтения более дорогой продукции).

# (6) Найти коэффициенты сезонности по месяцам.

```python
df['month'] = df['o_date'].dt.month
```

```python
season_pivot = df.pivot_table(index=['year', 'month'], values='price',aggfunc=np.sum)
season_pivot = season_pivot.reset_index()
season_pivot
```

```python
# mean_year_price = 
mean_year_price = season_pivot.pivot_table(index='year', values='price', aggfunc=np.mean)
mean_year_price
```

```python
season_pivot = season_pivot.merge(right=mean_year_price.reset_index(), left_on='year', right_on='year', how='left')
season_pivot
```

```python
season_pivot['seasonality'] = season_pivot['price_x'] / season_pivot['price_y']
season_pivot
```

```python
season_pivot['date'] = season_pivot['month'].astype(str) + '/' + season_pivot['year'].astype(str).str.slice(2)
sns.set_theme()
plt.figure(figsize=(12,8))
sns.lineplot(data=season_pivot, x='date', y='seasonality')
```
